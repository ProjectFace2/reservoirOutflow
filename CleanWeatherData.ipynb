```python
#Cleaning weather data year-wise(taluk level data-[temp and ])
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('/home/kishora/Downloads/Weather data.xlsx',sheet_name='2018')
```

```python
data['RECORDED_DATE'] = pd.to_datetime(data['RECORDED_DATE'],errors='coerce',dayfirst=True)
```

```python
wInput = pd.DataFrame()
data = data.set_index('RECORDED_DATE')
for group_name, df_group in data.groupby(pd.Grouper(freq='M')):
    #print(type(str(group_name)),"and then")
    me=str(group_name)
    if(me=='2018-06-30 00:00:00' or me=='2018-07-31 00:00:00' or me=='2018-08-31 00:00:00' or me=='2018-09-30 00:00:00' or me=='2018-10-31 00:00:00'):
        wInput = wInput.append(df_group)
```

```python
wInput
```

```python
weatherInput2017 = wInput
```

```python
weatherInput2017=weatherInput2017.drop(columns=['STATION_NUMBER'])
```

```python
weatherInput2017
```

```python
sep = {}
for group_name, df_group in weatherInput2017.groupby('DISTRICT'):
    sep[group_name] = df_group
```

```python
talukLevelData2018 = pd.DataFrame(columns = ['MINTEMPERATURE','MAXTEMPERATURE','MINHUMIDITY','MAXHUMIDITY','DISTRICT','TALUK_NAME'])
for group_name, df_group in sep.items():
    #sept[group_name] = df_group
    for name, df in df_group.groupby('TALUK_NAME'):
        grouped = df.groupby(pd.Grouper(freq='D')).mean()
        grouped['DISTRICT'] = group_name
        grouped['TALUK_NAME'] = name
        idx = pd.date_range('2018-06-01', '2018-10-31') 
        grouped=grouped.reindex(idx, fill_value=np.nan)
        grouped=grouped.replace({0:np.nan})
        grouped.interpolate(method='linear', inplace=True)
#         print(grouped)
        try:
            talukLevelData2018= pd.concat([talukLevelData2018,grouped])
        except:
            pass
#             talukLevelData2018.append(grouped,columns = ['MINTEMPERATURE','MAXTEMPERATURE','MINHUMIDITY','MAXHUMIDITY','DISTRICT','TALUK_NAME'])
print(talukLevelData2018)
```

```python
talukLevelData2018.to_csv('weatherdata2017.csv')
```
